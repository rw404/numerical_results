### Test 4

```python
def A_gen(i, j, m, n):
    if i == j:
        return n+m*m + (j/m) + (i/n)
    return (i+j)/(m+n)
def b_gen(i, m, n):
    return m*i+n
```

```python
n = 30
m = 20
A = np.random.random(size=[n,n])
b = np.random.random(size=n)
for i in range(n):
    for j in range(n):
        A[i][j] = A_gen(i, j, m, n)
    b[i] = b_gen(i, m, n)
```

```python
for x in np.linalg.solve(A, b):
    print("%lf" % x, end='\n')
print("Det", end=' ')
print(np.linalg.det(A))
print("A^-1", end='\n')
C = np.linalg.inv(A)
for i in range(n):
    for j in range(n):
        print("%lf" % C[i][j], end=' ')
    print("\n")
print("Condit", end=' ')
print(np.linalg.norm(A)*np.linalg.norm(np.linalg.inv(A)))
```
